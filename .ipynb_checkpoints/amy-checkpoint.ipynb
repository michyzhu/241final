```julia
using CSV
using LinearAlgebra
using Clustering
```

```julia
irisData = CSV.File("Iris Data Set.csv")
```

```julia
#generate similarity matrix
similarityMatrix = zeros(150, 150)
for row in irisData
    for col in irisData
        similarityMatrix[row.Index, col.Index] = abs(row.PetalLength - col.PetalLength);
    end
end
#display(similarityMatrix);
```

```julia
#generate sigma neighborhood adjacency matrix
adjacencyMatrix = zeros(150, 150)
for row in irisData
    for col in irisData
        if similarityMatrix[row.Index, col.Index] > .25
            adjacencyMatrix[row.Index, col.Index] = 0   
        else
            adjacencyMatrix[row.Index, col.Index] = 1
        end
        
    end
end
```

```julia
#generate degree matrix
degreeMatrix = zeros(150, 150)
for i = 1:150
    counter = 0
    for j = 1:150
        if i != j
            if adjacencyMatrix[i, j] == 1
                counter = counter + 1
            end
        end
    end
    degreeMatrix[i, i] = counter;
end
```

```julia
#generate laplacian matrix
laplacianMatrix = degreeMatrix - adjacencyMatrix
normalizedLaplacianMatrix = (degreeMatrix)^(1/2) * adjacencyMatrix * (degreeMatrix)^(1/2)
```

```julia
eigvecLaplacian = eigvecs(normalizedLaplacianMatrix)
```

```julia
#eigenvalue heuristic
eigvalLaplacian = eigvals(normalizedLaplacianMatrix)

maxGapRatio = -1
bestK = -1

for k = 2:10
    leftGap = abs(eigvalLaplacian[k] - eigvalLaplacian[k-1])
    rightGap = abs(eigvalLaplacian[k] - eigvalLaplacian[k+1])
    if (rightGap/leftGap > maxGapRatio)
        maxGapRatio = rightGap/leftGap
        bestK = k
    end
end

display(bestK) 
```

```julia
#get first k eigenvalues
kEigvecLaplacian = zeros(150, 3)
for i = 1:150
    for j = 1:3
        kEigvecLaplacian[i, j] = eigvecLaplacian[i, j]
    end
end
#display(kEigvecLaplacian)
```

```julia
X = transpose(kEigvecLaplacian)
R = kmeans(X, 3; maxiter=300, display=:iter)
scatter(iris.PetalLength, iris.PetalWidth, marker_z=R.assignments,
        color=:lightrainbow, legend=false)
```
